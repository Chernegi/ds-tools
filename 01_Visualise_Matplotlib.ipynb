## Matplotlib chart examples

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Linear Plot

```python
date = pd.date_range(start='2021-09-01', freq='D', periods=8)

fig, axs = plt.subplots()
axs.plot(date, [20, 21, 20, 19, 19, 20, 21, 20], linestyle="--",
         lw=1, marker='o', label='day temperature')
axs.plot(date, [15, 17, 16, 15, 15, 16, 15, 15], linestyle="--",
         lw=1, marker='o', label='night temperature')
plt.xlabel('Date', fontsize='small', fontstyle='italic', color='midnightblue')
plt.ylabel('Temperature', fontsize='small',
           fontstyle='italic', color='midnightblue')
plt.title('Day temperature. Kyiv', fontsize=15,
          fontstyle='italic', color='midnightblue', loc="center")
plt.text(date[0], 15, 'Autumn has come', color="blue")
plt.xticks(fontsize=6)
plt.yticks(fontsize=6)
plt.legend(loc="upper right")
plt.show()
```

## Split plots

```python
fig, axs = plt.subplots(2, 1)

axs[0].plot(date, [20, 21, 20, 19, 19, 20, 21, 20], label='day temperature')
axs[1].plot(date, [15, 17, 16, 15, 15, 16, 15, 15], label='night temperature')

axs[0].set_title('Day temperature', fontsize=10)
axs[1].set_title('Night Temperature', fontsize=10)

fig.suptitle('Temperature. Kyiv', fontsize=15)
plt.show()
```

## Histograms with matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import colors
from matplotlib.ticker import PercentFormatter

# Create a random number generator with a fixed seed for reproducibility
rng = np.random.default_rng(19680801)

N_points = 100000
n_bins = 50

# Generate two normal distributions
dist1 = rng.standard_normal(N_points)
dist2 = 0.4 * rng.standard_normal(N_points) + 5

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

# We can set the number of bins with the *bins* keyword argument.
axs[0].hist(dist1, bins=n_bins)
axs[1].hist(dist2, bins=n_bins)
```

## Polar Coordinates

```python

import matplotlib.pyplot as plt
import numpy as np

phase = np.linspace(0, 2.0 * np.pi, 1000)

r = np.sin(3 * phase)
plt.polar(phase, r, c="r")
plt.show()
```
