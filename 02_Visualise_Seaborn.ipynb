```python
import ssl
import seaborn as sns
import pandas as pd

# to avoid SSLCertVerificationError while loading dataset
ssl._create_default_https_context = ssl._create_unverified_context
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
# load builtin dataset "miles per galon"

data = sns.load_dataset("mpg")
data['origin'] = data['origin'].astype('category')
data.head(6)
```

```python
sns.set_style("darkgrid")
sns.lineplot(x='model_year', y='horsepower', hue='origin', data=data)
```

```python
# scatterplot
sns.scatterplot(x='mpg', y='displacement', hue='origin', data=data)
```

```python
sns.relplot(x='mpg', y='displacement', kind='scatter',
            hue='origin', col='origin', data=data)
```

### Categorical data visualisation

```python
sns.stripplot(x='origin', y='mpg', data=data)
```

```python
# mark numder of cylinders with color
sns.stripplot(x='origin', y='mpg', hue='cylinders', data=data)
```

```python
# swarmplot - cateegories are placesed in line
sns.swarmplot(x='origin', y='mpg', data=data)
```

```python
# boxplot (або японські свічки)
sns.boxplot(x="origin", y="mpg", hue="cylinders", data=data)
```

```python
# violinplot shows dencity
sns.violinplot(x="origin", y="mpg", data=data)
```

```python
# boxenplot also shows dencity
sns.boxenplot(x="origin", y="mpg", data=data)
```

```python
# countplot
sns.countplot(x="cylinders", data=data)
```

```python
# linear regression visualisation out of the box
sns.lmplot(x="horsepower", y="displacement", data=data)
```
