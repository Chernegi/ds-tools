де x=[x1,x2,...xd]  - вектор ознак об'єкта, σi та f ij - деякі неперервні функції, причому f ij не залежить від вибору a(x).
Функція a(x) визначена лише на одиничному кубі, а отже, всі елементи вибірки мають лежати у ньому. Це не є проблемою: завжди можна відмасштабувати ознаки так, щоб на навчальній вибірці кожна ознака приймала значення з відрізка [0,1].
